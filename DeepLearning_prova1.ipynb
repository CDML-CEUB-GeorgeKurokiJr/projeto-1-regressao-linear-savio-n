```python
# =====================================
# 1. BAIXAR DATASET DA UCI
# =====================================
!wget -O gas.zip "https://archive.ics.uci.edu/static/public/322/gas+sensor+array+under+dynamic+gas+mixtures.zip"

# =====================================
# 2. DESCOMPACTAR
# =====================================
import zipfile
import os

with zipfile.ZipFile("gas.zip", 'r') as zip_ref:
    zip_ref.extractall("gas_dataset")

print("Arquivos extraídos:")
print(os.listdir("gas_dataset"))

# =====================================
# 3. CARREGAR ARQUIVOS TXT COM PANDAS (CORRIGIDO)
# =====================================

import pandas as pd

path_co = "gas_dataset/ethylene_CO.txt"
path_methane = "gas_dataset/ethylene_methane.txt"

print("Carregando...")

# nomes corretos das colunas
colunas = (
    ["time", "gas_conc", "ethylene_conc"] +
    [f"sensor_{i}" for i in range(1, 17)]
)

# carregar ignorando linha de descrição textual
df_co = pd.read_csv(
    path_co,
    sep=r"\s+",
    skiprows=1,
    header=None,
    names=colunas
)

df_methane = pd.read_csv(
    path_methane,
    sep=r"\s+",
    skiprows=1,
    header=None,
    names=colunas
)

# =====================================
# MOSTRAR INFORMAÇÕES
# =====================================

print("===== ETHYLENE CO =====")

print("\nHEAD:")
display(df_co.head())

print("\nSHAPE:")
print(df_co.shape)

print("\nCOUNT (valores nao nulos por coluna):")
print(df_co.count())


print("\n\n===== ETHYLENE METHANE =====")

print("\nHEAD:")
display(df_methane.head())

print("\nSHAPE:")
print(df_methane.shape)

print("\nCOUNT (valores nao nulos por coluna):")
print(df_methane.count())
```

# **regressao em si**

```python
# =====================================
# PREPARAR DADOS PARA PYTORCH
# =====================================

import torch
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# escolher dataset (use CO primeiro)
df = df_co.copy()

# --------------------------
# FEATURES E TARGET
# --------------------------

X = df[[f"sensor_{i}" for i in range(1, 17)]]
y = df["gas_conc"]

print("Instancias:", len(df))
print("Features:", X.shape[1])

# --------------------------
# NORMALIZAR (ESSENCIAL)
# --------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --------------------------
# TRAIN TEST SPLIT
# --------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# --------------------------
# CONVERTER PARA TENSOR
# --------------------------

X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)

y_train = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
y_test = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)

print("Tensor treino:", X_train.shape)
print("Tensor teste:", X_test.shape)
```

```python
# =====================================
# REGRESSAO LINEAR PYTORCH
# =====================================

import torch.nn as nn

# --------------------------
# MODELO
# --------------------------

class LinearRegressionModel(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.linear = nn.Linear(n_features, 1)

    def forward(self, x):
        return self.linear(x)

model = LinearRegressionModel(X_train.shape[1])

# --------------------------
# LOSS E OTIMIZADOR
# --------------------------

criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# --------------------------
# TRAINING LOOP
# --------------------------

epochs = 10

for epoch in range(epochs):

    predictions = model(X_train)
    loss = criterion(predictions, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch+1}/{epochs} | Loss: {loss.item():.6f}")

# --------------------------
# AVALIAÇÃO
# --------------------------

with torch.no_grad():
    test_pred = model(X_test)
    test_loss = criterion(test_pred, y_test)

print("\nTest Loss:", test_loss.item())
```

rmse

```python
# --------------------------
# AVALIAÇÃO
# --------------------------

with torch.no_grad():
    test_pred = model(X_test)
    test_loss = criterion(test_pred, y_test)

    # RMSE
    test_rmse = torch.sqrt(test_loss)

print("\nTest MSE:", test_loss.item())
print("Test RMSE:", test_rmse.item())



with torch.no_grad():
    train_pred = model(X_train)
    train_mse = criterion(train_pred, y_train)
    train_rmse = torch.sqrt(train_mse)

print("Train RMSE:", train_rmse.item())
```
